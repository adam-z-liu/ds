```python
%matplotlib inline

import pandas as pd
import numpy as np
import seaborn as sns

import pylab
pylab.rcParams['figure.figsize'] = (30.0, 15.0) # 显示大小
```

```python
data_bigquery_path = './data/bigquery.csv'
data_tmdb_path = "./data/tmdb.csv"

data_bigquery = pd.read_csv(data_bigquery_path)
data_tmdb = pd.read_csv(data_tmdb_path)
```

```python
data_movie = data_tmdb[data_tmdb['topn']==0][['MDS_SeriesName','series_name','content_length_ms']]
mergedf = pd.merge(data_bigquery, data_movie, how='left', left_on='asset_seriesName', right_on='MDS_SeriesName')
outdf = mergedf[mergedf['asset_seriesName'] == mergedf['series_name']].drop_duplicates() 
outdf = outdf[outdf.content_length_ms.notnull()]
outdf
```
